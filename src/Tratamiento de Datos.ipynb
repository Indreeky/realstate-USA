```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time

# Configura las opciones del navegador
chrome_options = Options()
chrome_options.add_argument('--headless')  # Opcional, para ejecutar en modo sin cabeza

# Ruta a Chromedriver
service = Service('G:\Mi unidad\Data Science\proyecto\Proyecto-predictor-de-precios-de-propiedades-en-EEUU\src\chromedriver.exe')

# Inicializa el navegador
driver = webdriver.Chrome(service=service, options=chrome_options)

# Lista de estados
states =  [
    '01', '05', '09', '10', '19', '23',
    '28', '30', '33', '38',
    '44', '45', '46', '49', 
    '50', '54', '56'
]

# Crear un DataFrame vacío
columns = ["state", "columna14", "one_adult_no_kids_living_wage",
    "one_adult_one_kid_living_wage",
    "one_adult_two_kids_living_wage",
    "one_adult_three_kids_living_wage",
    "two_adults_one_working_no_kids_living_wage",
    "two_adults_one_working_one_kid_living_wage",
    "two_adults_one_working_two_kids_living_wage",
    "two_adults_one_working_three_kids_living_wage",
    "two_adults_both_working_no_kids_living_wage",
    "two_adults_both_working_one_kid_living_wage",
    "two_adults_both_working_two_kids_living_wage",
    "two_adults_both_working_three_kids_living_wage"
]
data = []

try:
    for state in states:
        # Construye la URL para el estado actual
        url = f"https://livingwage.mit.edu/states/{state}"
        

        # Abre la página correspondiente
        driver.get(url)

        # Espera un tiempo para asegurar que la página cargue
        time.sleep(2)

        # Encuentra la tabla de "living wage"
        tables = driver.find_elements(By.TAG_NAME, 'table')

        if tables:
            # La primera tabla generalmente contiene los "living wage"
            rows = tables[0].find_elements(By.TAG_NAME, 'tr')

            # Extrae los datos de la primera fila relevante
            if len(rows) > 1:
                wage_data = rows[2].find_elements(By.TAG_NAME, 'td')
                data_row = [state] + [cell.text for cell in wage_data]

                # Verifica la longitud de los datos extraídos
                if len(data_row) == len(columns):
                    data.append(data_row)
                else:
                    print(f"Longitud incorrecta para el estado {state}: {len(data_row)} columnas detectadas.")

finally:
    # Cierra el navegador
    driver.quit()

# Crear DataFrame
df_aux = pd.DataFrame(data, columns=columns)

df_aux.head()
```

```python
import numpy as np

# Crear una nueva columna para almacenar price/house_size en pies cuadrados
df['price_per_house_size'] = df['price'] / df['house_size']

# Calcular el promedio de price_per_house_size por cada zip_code, ignorando nulos
promedios_zip_code = df.groupby('zip_code')['price_per_house_size'].mean()

promedios_zip_code
```

```python
# Función para rellenar valores nulos en house_size usando el promedio calculado
def llenar_house_size(row):
    if np.isnan(row['house_size']):
        promedio = promedios_zip_code.get(row['zip_code'], np.nan)
        return row['price'] / promedio if promedio else np.nan
    else:
        return row['house_size']

# Aplicar la función para llenar valores nulos en house_size
df['house_size'] = df.apply(llenar_house_size, axis=1)

# Eliminar la columna intermedia utilizada para el cálculo
df.drop(columns=['price_per_house_size'], inplace=True)
```

```python
# Función para asignar 'bed' y 'bath' basado en 'house_size' en pies cuadrados
def asignar_bed_bath(row):
    if np.isnan(row['bed']) or np.isnan(row['bath']):
        if row['house_size'] < 1000:
            return 1, 1  # 1 cama, 1 baño
        elif 1000 <= row['house_size'] < 2000:
            return 2, 1  # 2 camas, 1 baño
        elif 2000 <= row['house_size'] < 3000:
            return 3, 2  # 3 camas, 2 baños
        else:
            return 4, 3  # 4 camas, 3 baños
    else:
        return row['bed'], row['bath']

# Aplicar la función y crear nuevas columnas para 'bed' y 'bath'
df[['bed', 'bath']] = df.apply(asignar_bed_bath, axis=1, result_type='expand')

# Mostrar el DataFrame actualizado
df
```
