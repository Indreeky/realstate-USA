**DEFINIR VARIANTES NUMÉRICAS**

```python
# Gráfico de barras para "bed"
plt.figure(figsize=(7, 5))
df["bed"].value_counts().plot(kind="bar", color="lightgreen", edgecolor="k")
plt.title('Bed distribution')
plt.xlabel('Bed')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Boxplot (Diagrama de Caja y Bigotes):outliers y ver la mediana, cuartiles y rango intercuartílico.
plt.figure(figsize=(8, 6))
plt.boxplot(df['bed'], vert=False)
plt.title('Boxplot de Habitaciones')
plt.xlabel('bed')
plt.show()
```

```python
# Calcula los cuartiles y el IQR
Q1 = df['bed'].quantile(0.25)
Q3 = df['bed'].quantile(0.75)
IQR = Q3 - Q1

# Define los límites inferior y superior para los outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifica los outliers
outliers = df[(df['bed'] < lower_bound) | (df['bed'] > upper_bound)]

# Muestra los outliers
print(f"Outliers encontrados: {outliers.shape[0]}")
print(f"lower_bound: {lower_bound}")
print(f"Upper_bound: {upper_bound}")
outliers
```

```python
# Violin Plot: densidad de los datos a diferentes valores, proporcionando más información sobre la distribución

plt.figure(figsize=(8, 6))
sns.violinplot(x=df['bed'])
plt.title('Violin Plot de Habitaciones')
plt.xlabel('bed')
plt.show()
```

```python
# Filtra el DataFrame para eliminar los outliers
df_clean = df[(df['bed'] >= lower_bound) & (df['bed'] <= upper_bound)]

# Verifica el nuevo tamaño del DataFrame
df_clean.shape
```

```python
df = df_clean
```

### Análisis del Gráfico ###

**Distribución de Precios por Estado**

En general, la mayoría de los estados tienen precios que se agrupan en valores relativamente bajos, pero hay algunos estados que muestran una variabilidad significativa con algunos precios extremadamente altos, que se representan como puntos fuera de los límites de las "cajas" (outliers).
Outliers:

Algunos estados presentan outliers muy marcados, con precios considerablemente más altos que la mediana de los precios en ese estado. **Mediana y Rango Intercuartílico (IQR)**

Las cajas representan el rango intercuartílico (IQR), es decir, la diferencia entre el primer cuartil (Q1) y el tercer cuartil (Q3), que abarca el 50% de los datos.
La línea dentro de la caja representa la mediana (Q2), que es el valor medio de los precios en cada estado.
En la mayoría de los estados, la mediana parece estar cercana a la base de la caja, lo que indica que la mayoría de los precios están concentrados en el cuartil inferior del rango.
Variabilidad entre Estados:

Estados como Massachusetts, California y Florida muestran una mayor dispersión de precios, lo que indica una alta variabilidad en los precios de las propiedades en estos estados.
Por otro lado, estados como Alaska, Idaho y Wyoming parecen tener una distribución de precios más concentrada y sin grandes outliers, lo que sugiere una menor variabilidad en los precios.
Escala del Gráfico:

La escala del eje Y es bastante alta, alcanzando hasta valores de 1e8 (100 millones), lo que indica que algunos inmuebles pueden estar valorados en decenas o cientos de millones de dólares.
La escala grande puede hacer que las diferencias en los precios de la mayoría de los estados parezcan pequeñas, aunque en realidad puede haber diferencias significativas dentro de un rango más común (por ejemplo, menos de 10 millones de dólares).

**Observaciones Generales**

El gráfico sugiere que, aunque en la mayoría de los estados los precios de los inmuebles tienden a estar dentro de un rango moderado, existen excepciones significativas en algunos estados, donde los precios pueden llegar a ser extremadamente altos. Esto podría reflejar las diferencias en los mercados inmobiliarios de lujo o en zonas con alta demanda.
